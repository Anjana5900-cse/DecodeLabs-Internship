```python
# ==========================================
# DecodeLabs - Project 3: Industrial Training Kit
# Capstone: Tech Stack Recommender System
# ==========================================

import os
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

dataset_path = "raw_skills.csv"

if not os.path.exists(dataset_path):
    print(f"⚠️ '{dataset_path}' not detected in session storage.")
    print("✨ Generating the official skill-to-role database template...")

    mock_data = {
        "Job Role": [
            "Data Scientist",
            "DevOps Engineer",
            "Backend Developer",
            "Cloud Architect",
            "Sys Admin"
        ],
        "Required Skills": [
            "Python SQL Machine Learning Data Analysis Deep Learning NumPy Pandas",
            "AWS Docker Kubernetes Git CI/CD Cloud Computing Automation Jenkins",
            "Java Python SQL APIs Backend Development Django Flask SpringBoot",
            "Cloud Computing AWS Azure Infrastructure Automation Cloud Architect",
            "Linux Networking Automation Bash Security Scripting System Admin"
        ]
    }
    pd.DataFrame(mock_data).to_csv(dataset_path, index=False)
    print("✅ 'raw_skills.csv' is now ready inside your session!\n")

def run_matchmaker_engine(user_skills):

    df = pd.read_csv(dataset_path)
    df = df.dropna(subset=["Job Role", "Required Skills"])

    user_input_str = " ".join(user_skills)
    all_documents = df["Required Skills"].tolist() + [user_input_str]


    vectorizer = TfidfVectorizer()
    tfidf_matrix = vectorizer.fit_transform(all_documents)


    job_vectors = tfidf_matrix[:-1]
    user_vector = tfidf_matrix[-1:]


    similarity_scores = cosine_similarity(user_vector, job_vectors).flatten()
    df["Similarity Score"] = similarity_scores


    top_3_recommendations = df.sort_values(by="Similarity Score", ascending=False).head(3)


    print("\n" + "="*50)
    print("         DIGITAL MATCHMAKER OUTPUT ENGINE         ")
    print("="*50)
    print(f"📥 Captured User State: {user_skills}\n")
    print("⚙️ Computing similarity metrics across vector space...\n")
    print("--- 🎯 Top 3 Recommended Career Paths ---")

    for index, row in top_3_recommendations.iterrows():
        if row["Similarity Score"] > 0:
            match_percentage = row["Similarity Score"] * 100
            print(f"• Role: {row['Job Role']:<20} | Angular Alignment Match: {match_percentage:.2f}%")
        else:
            print("❌ No matching item attributes found for this skill matrix row.")
            break
    print("="*50 + "\n")



Enter_Your_Skills = "Python, Cloud Computing, Automation" # @param {type:"string"}

parsed_skills = [skill.strip() for skill in Enter_Your_Skills.split(",") if skill.strip()]

if parsed_skills:
    run_matchmaker_engine(parsed_skills)
else:
    print("🛑 Input field empty! Please enter skill keywords separated by commas.")
```
